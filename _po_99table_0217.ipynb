```python
# 九九乘法表
```

```python
# 1. 直覺這樣寫
```

```python
for i in range(1, 10):
    for j in range(1, 10):
        print(f"{i}*{j}={i*j:2}")  # print出"f字串" <- 關鍵字，字串的格式整理有format, f字串...方法，可以估狗學習更多
```

```python
# 2.
# 執行後觀察可以列出全部運算，但是不是目標的表格樣子
# 所以調整一下:
```

```python
for i in range(1, 10):
    for j in range(1, 10):
        print(f"{i}*{j}={i*j}", end="  ")  
    print()
```

```python
# 3.
# 樣子對了，但是被乘數和成數反過來了，一開始邏輯錯了
```

```python
for i in range(1, 10):
    for j in range(1, 10):
        print(f"{j}*{i}={j*i}", end="  ")  
    print()
```

```python
# 4.
# 不夠漂亮，再讓它更像表格一些
```

```python
print("\n－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－")
for i in range(1, 10):
    for j in range(1, 10):
        print(f"| {j}*{i}={j*i:2}", end=" | ")  # f字串的數字位數對齊
    print("\n－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－")
```

```python
# 5.
# 讓程式碼更整潔
# 把重複寫兩字的"值"，用一個變數來使用，如此一來就可以重複使用
```

```python
# 宣告字串型態的變數: str_prt_vertical_line
str_prt_vertical_line = "\n－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－－"

print(str_prt_vertical_line)  # 印出該變數
for i in range(1, 10):
    for j in range(1, 10):
        print(f"| {j}*{i}={j*i:2}", end=" | ")  # f字串的數字位數對齊
    print(str_prt_vertical_line)  # 印出該變數
```
